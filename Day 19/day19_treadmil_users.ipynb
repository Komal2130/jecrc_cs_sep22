# MultiVariate Analysis

```python
sns.heatmap(
    data = data[num_cols].corr() ,
    annot=True ,
    cmap = choice(p_list)
)
```

```python
for col in cat_cols:
    if col == 'Product':
        continue
    for ncol in num_cols:
        plt.figure(figsize=(10,3))
        plt.suptitle(f'{col},vs,{ncol}')
        plt.subplot(1,3,1)
        sns.boxplot(
            data = data, x= col,y = ncol,
            hue = 'Product', palette=choice(p_list))
        plt.subplot(1,3,2)
        sns.violinplot(
            data = data, x= col,y = ncol, 
            hue = 'Product',palette=choice(p_list))
        plt.subplot(1,3,3)
        sns.barplot(
            data = data, x= col,y = ncol,
            hue = 'Product', palette=choice(p_list))
        plt.show()
```
