# Types of Graphs
1. plot -          line graph between two parameters
2. bar -           bar graph of single parameters (optional cat_data)
3. hist -          histogram of a bunch of data i.e. frequency of single parameters
4. scatter -       dot graph between two parameters
5. boxplot / box - statistical overview in form of box for a single parameters
                   mean, outliers, qunatiles(25%, 75%)
                   (optional - categorical data)
6. stem -          more like bar graph , lines instead of Bars 
7. violinplot -    Violin shaped statistical overview graph 
                   spread of data(more like hist) 
                   single parameter (optional cat_data)

```python
plt.bar(x,y)
plt.title('Bar Graph')
plt.show()

plt.hist(x,y)
plt.title('Histogram')
plt.show()

plt.scatter(x,y)
plt.title('Scatter')
plt.grid()
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()

plt.boxplot(x)
plt.title('Boxplot')
plt.show()

plt.stem(x,y)
plt.title('Stem')
plt.show()

plt.violinplot(x)
plt.title('violinplot')
plt.show()

```

```python
fig = plt.figure()
plt.plot(x,y,lw=3,marker='o', markersize = 15,markerfacecolor='purple')
plt.plot(x,y2, marker = 'v', markersize = 10)
plt.legend(('y=x**3','y=x**4'))
plt.grid()
plt.show()
```
